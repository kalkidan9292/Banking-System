```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Banking.csv')
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
#Generate descriptive statistics for the dataframe
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low','Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#Examine teh distruction of unique categories in categorical columns
categorical_cols = df[["BRId", "GenderId","IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Properties Owned", "Risk Weighting", "Income Band"]]
for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

## Univariate Analysis

```python
for i, predictor in enumerate(df):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

## Bivariate Analysis

```python
for i, predictor in enumerate(["BRId", "GenderId", "IAId", "Amount of Credit Cards",
                               "Nationality", "Occupation", "Fee Structure",
                               "Loyalty Classification", "Properties Owned",
                               "Risk Weighting", "Income Band"]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# Histplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income','Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

## Heatmaps


```python
numerical_cols = ['Estimated Income','Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Correlation Heatmap Matrix')
plt.show()
```
